```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.keys import Keys
import pandas as pd
import tqdm
from pytube import YouTube
```

```python
driver = webdriver.Edge(r'C:\\Users\\sarkh\\Downloads\\edgedriver_win64\\msedgedriver.exe')
driver.get('https://www.google.com')
```

```python
driver.get('https://www.youtube.com/')
searchbar = driver.find_element(by=By.CSS_SELECTOR, value="input#search")
searchbar.send_keys("advertisement")
searchbar.send_keys(Keys.ENTER)

videos = driver.find_elements(by=By.CSS_SELECTOR, value="ytd-video-renderer")
while len(videos) < 50:
    driver.find_element(by=By.TAG_NAME, value="body").send_keys(Keys.END)
    videos = driver.find_elements(by=By.CSS_SELECTOR, value="ytd-video-renderer")
```

```python
data = []
for video in tqdm.tqdm(videos):
    d = {}
    d['Title'] = video.find_element(by=By.CSS_SELECTOR, value="a#video-title")
    d['URL'] = '' if d['Title'] is None else d['Title'].get_attribute('href')
    d['Title'] = '' if d['Title'] is None else d['Title'].get_attribute('title')
    
    d['Channel'] = video.find_element(by=By.CSS_SELECTOR, 
                                    value="*#channel-name a")
    d['Channel'] = None if d['Channel'] is None else d['Channel'].get_attribute('href')
    link = YouTube(d['URL'])
    d['Views'] = 0 if link == '' else link.views
    d['Duration in s'] = 0 if link == '' else link.length
    """
    d['Vues'] = video.find_element(by=By.CSS_SELECTOR, 
                                    value="div#metadata-line span:nth-child(1)")
    d['Vues'] = '' if d['Vues'] is None else d['Vues'].text
    
    d['Posté le'] = video.find_element(by=By.CSS_SELECTOR, 
                                    value="div#metadata-line span:nth-child(2)")
    d['Posté le'] = '' if d['Posté le'] is None else d['Posté le'].text
    """
    
    data.append(d)

    
df = pd.DataFrame(data)
```
